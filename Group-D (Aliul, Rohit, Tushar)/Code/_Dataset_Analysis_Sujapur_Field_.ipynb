***Dataset Analysis @Area_02_ Sujapur_Field***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('/content/air_sujapurfield_avg.csv')
```

```python
print (data)
```

```python
time = data.iloc[2:, 0]  # Assuming the time starts from the third row
co2 = data.iloc[2:, 1].astype(float)
toluene = data.iloc[2:, 2].astype(float)
nh4 = data.iloc[2:, 3].astype(float)
acetone = data.iloc[2:, 4].astype(float)
co = data.iloc[2:, 5].astype(float)
```

```python
# Create a line plot for CO2
plt.plot(time, co2, label='CO2')

# Create a line plot for Toluene
plt.plot(time, toluene, label='Toluene')

# Create a line plot for NH4
plt.plot(time, nh4, label='NH4')

# Create a line plot for Acetone
plt.plot(time, acetone, label='Acetone')

# Create a line plot for CO
plt.plot(time, co, label='CO')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Gas Levels')
plt.title('Gas Levels over Time @SUJAPUR_FIELD')

# Rotate x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Show the line plot
plt.show()
```
